```python
%matplotlib inline
import pandas as pd
import matplotlib as mp
import matplotlib.pyplot as plt
import seaborn as sns
from statistics import stdev 
```

```python
sns.set(style='whitegrid')
#color_codes=True
#"whitegrid"  "dark"  "ticks"  "white"

data="DataD.xlsx"

df = pd.read_excel(data)
df.head()
```

```python
def ticks(LL, UL, step):
    lst=[round(i*step,4)+LL for i in range(0,int((UL-LL)/step+0.5))]
    lst.append(UL)
    return lst
```

```python
import pylab as pl
plt.figure(figsize=(37,30))

pal='flare'
#w=sns.scatterplot(data=df,x=list(df)[1],y=list(df)[2],s=100,c=df['Z'],vmin=-0.03,vmax=0.03,cmap='coolwarm')
plt.scatter(x=df['X'],y=df['Y'],c=df['Z'],vmin=-0.03,vmax=0.03,cmap='coolwarm',s=200)

plt.xticks(fontsize=24,ticks=ticks(-25,25,5))
plt.yticks(fontsize=24,ticks=ticks(-25,25,5))



cbar = plt.colorbar()
for t in cbar.ax.get_yticklabels():
     t.set_fontsize(24)
```
